```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def plot_3D(x, y, u):
    '''Creates 3D plot with appropriate limits and viewing angle
    
    Parameters:
    ----------
    x: array of float
        nodal coordinates in x
    y: array of float
        nodal coordinates in y
    u: 2D array of float
        calculated potential field
    
    '''
    fig = pyplot.figure(figsize=(11,7), dpi=100)
    ax = fig.gca(projection='3d')
    X,Y = numpy.meshgrid(x,y)
    surf = ax.plot_surface(X,Y,u[:], rstride=1, cstride=1, cmap=cm.viridis,
            linewidth=0, antialiased=False)

    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')
    ax.view_init(30,45)
```

```python
def poisson_jacobi(u,f,dx,dy,l2_target,n):
    l2_norm = 5
    iterations = 0
    l2_conv=[]
    
    while l2_norm>l2_target:
        un = u.copy()
        u[1:-1,1:-1] =  0.25*(un[1:-1,:-2]+un[1:-1,2:]+un[:-2,1:-1]+un[2:,1:-1]-dx**2*f[1:-1,1:-1])
        
        #neumann bc
        u[1:-1,0] = 0.25*(2*un[1:-1,1]- 2*un[1:-1,0]+un[:-2,0]+un[2:,0]-dx**2*f[1:-1,0])
        u[1:-1,-1]= 0.25*(2*un[1:-1,-2]- 2*un[1:-1,-1]+un[:-2,-1]+un[2:,-1]-dx**2*f[1:-1,-1])
        u[0,1:-1] = 0.25*(2*un[1,1:-1]- 2*un[0,1:-1]+un[0,:-2]+un[0,2:]-dx**2*f[0,1:-1])
        u[-1,1:-1]= 0.25*(2*un[-2,1:-1]- 2*un[-1,1:-1]+un[-1,:-2]+un[-1,2:]-dx**2*f[-1,1:-1])
        
        u[0,0] = (un[0,1]+un[1,0])/2
        u[0,-1] = (un[0,-2]+un[1,-1])/2
        u[-1,0] = (un[-2,0]+un[-1,1])/2
        u[-1,-1] = (un[-1,-2]+un[-2,-1])/2
        
        l2_norm = L2_error(u,1)
        iterations+=1
        l2_conv.append(l2_norm)
        
        print("Jacobi method took {} iterations at tolerance {} with nx = {}".format(iterations,l2_target,nx))
        return u,l2_conv
```

```python
def u_exact(n,X,Y):
    return numpy.cos(2*numpy.pi*n*X)*numpy.cos(2*numpy.pi*n*Y)

def L2_error(u,un):
    return numpy.sqrt(numpy.sum((u-un)**2)/numpy.sum(un**2))
```
