```python
import numpy
import sympy
```

```python
a,b,c,d = sympy.symbols("a b c d")
z = sympy.Matrix(([[1,1,1,1,1,] , [0, 1*a, 2*b, 3*c, 4*d],[0, 0.5*a**2, 2*b**2, (9/2)*c**2, 8*d**2],[0, (1/6)*a**3, (8/6)*b**3, (27/6)*c**3, (64/6)*d**3],\
    [0, (1/24)*a**4, (16/24)*b**4, (81/24)*c**4, (256/24)*d**4]]))
print(z)
```

```python
A_1 =z.inv()
```

```python
ff = sympy.Matrix(([[0],[1],[0],[0],[0]]))
```

```python
pfft = A_1*ff
```

```python
print (pfft)
```

```python
new = sympy.Matrix(([[1,2,3,4],[0.5,2,(9/2),8],[(1/6),(8/6),(27/6),(64/6)],[(1/24),(16/24),(81/24),(256/24)]]))
fnew =  sympy.Matrix(([[1],[0],[0],[0]]))

new_inv = new.inv()
```
