```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

# Jacobi w/ dirichlet bc

```python
def plot_3D(x, y, u):
    '''Creates 3D plot with appropriate limits and viewing angle
    
    Parameters:
    ----------
    x: array of float
        nodal coordinates in x
    y: array of float
        nodal coordinates in y
    u: 2D array of float
        calculated potential field
    
    '''
    fig = pyplot.figure(figsize=(11,7), dpi=100)
    ax = fig.gca(projection='3d')
    X,Y = numpy.meshgrid(x,y)
    surf = ax.plot_surface(X,Y,u[:], rstride=1, cstride=1, cmap=cm.viridis,
            linewidth=0, antialiased=False)

    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_xlabel('$x$')
    ax.set_ylabel('$y$')
    ax.set_zlabel('$z$')
    ax.view_init(30,45)
```

```python
def poisson_IG(n,nx):
    ny=nx
    dx = (1)/(nx-1)
    dy = (1)/(ny-1)
    
    x = numpy.linspace(0,1,nx)
    y = numpy.linspace(0,1,ny)
    X,Y=numpy.meshgrid(x,y)
    
    f = -8*numpy.pi**2*n**2*numpy.sin(2*numpy.pi*n*Y)*numpy.sin(2*numpy.pi*n*X)
    
    u_i = numpy.zeros((ny,nx))    #initial guess
    
    return X,Y,x,y,u_i,f,dx,dy
```

```python
def poisson_jacobi(u,f,dx,dy,l2_target):
    l2_norm = 1
    iterations = 0
    l2_conv=[]
    
    while l2_norm > l2_target:
        ud = u.copy()
        #discretization with jacobi relaxation
        u[1:-1,1:-1] = 1/(2*(dx**2 + dy**2)) * \
                        ((ud[1:-1,2:]+ud[1:-1,:-2])*dy**2 +\
                        (ud[2:,1:-1] + ud[:-2,1:-1])*dx**2 -\
                         f[1:-1,1:-1]*dx**2*dy**2)
        
        #dont need to add dirichlet- auto enforced
        
        l2_norm = l2_error(u,ud)
        iterations+=1
        l2_conv.append(l2_norm)
        
    print("Jacobi method took {} iterations\
 at tolerance {} with nx = {}".format(iterations, l2_target,nx))
        
    return u,l2_conv        
```

```python
def u_exact(n,X,Y):
    return numpy.sin(2*numpy.pi*n*X)*numpy.sin(2*numpy.pi*n*Y)
```

```python
def l2_error(u,u_e):
    return numpy.sqrt(numpy.sum((u-u_e)**2)/numpy.sum(u_e**2))

def L2_error(p, pn):
    return numpy.sqrt(numpy.sum((p - pn)**2)/numpy.sum(pn**2))
```

# (a) Jacobi
# n = 1

```python
#initalize
nx = 81
ny = 81
n1 = 1    #n = (nx-1)/2 range for n values
X,Y,x,y,u_i,f,dx,dy = poisson_IG(n1,nx)
```

```python
#check initial guess
plot_3D(x,y,u_i)
```

```python
l2_target = 1e-8

u, l2_conv1 = poisson_jacobi(u_i.copy(),f,dx,dy,l2_target)
```

```python
plot_3D(x,y,u)
u_e= u_exact(n1,X,Y)
numpy.max(u)
```

```python
error=L2_error(u,u_e)
error
```

```python
pyplot.semilogy(numpy.arange(len(l2_conv1)),l2_conv1,lw=2)
pyplot.xlabel('iterations')
pyplot.ylabel('convergence')
pyplot.title('Iterations until Convergence, Jacobi, n=1');
```

```python
nx_val = [11,21,41,81,101]    #increase nx to see how error behaves with increased number of points
error_j_n1 = numpy.zeros_like(nx_val,dtype=numpy.float)
for i, nx in enumerate(nx_val):
    ny = nx
    X,Y,x,y,u_i,f,dx,dy=poisson_IG(n1,nx)
    u,l2_conv = poisson_jacobi(u_i.copy(),f,dx,dy,l2_target)
    u_e = u_exact(n1,X,Y)
    error_j_n1[i] = L2_error(u,u_e)
    plot_3D(x,y,u)
pyplot.figure(figsize=(6,6))
pyplot.grid(True)
pyplot.xlabel(r'$n_x$', fontsize=18)
pyplot.ylabel(r'$L_2$-norm of the error', fontsize=18)
pyplot.title('Error vs number of points on grid, Jacobi, n=1', fontsize = 18)
pyplot.loglog(nx_val, error_j_n1, color='k', ls='--', lw=2, marker='o')
pyplot.axis('equal');
```

# n = 2

```python
#initalize
nx = 81
ny = 81
n2 = 2    #n = (min(nx)-1)/2 range for n values, min nx = 11
X,Y,x,y,u_i,f,dx,dy = poisson_IG(n2,nx)

#check initial guess
plot_3D(x,y,u_i)

# set tolerance
l2_target = 1e-8

#iterate u values at each point on mesh grid
u, l2_conv2 = poisson_jacobi(u_i.copy(),f,dx,dy,l2_target)

#check final u
plot_3D(x,y,u)

#calculate exact u
u_e= u_exact(n2,X,Y)

#find error between calculated u and exact u
error=L2_error(u,u_e)
error
```

```python
#plot number of iterations until convergence
pyplot.semilogy(numpy.arange(len(l2_conv2)),l2_conv2,lw=2)
pyplot.xlabel('iterations')
pyplot.ylabel('convergence')
pyplot.title('Iterations until Convergence, Jacobi, n=2');
```

```python
nx_val = [11,21,41,81,101]    #increase nx to see how error behaves with increased number of points
error_j_n2 = numpy.zeros_like(nx_val,dtype=numpy.float)
for i, nx in enumerate(nx_val):
    ny = nx
    X,Y,x,y,u_i,f,dx,dy=poisson_IG(n2,nx)
    u,l2_conv = poisson_jacobi(u_i.copy(),f,dx,dy,l2_target)
    u_e = u_exact(n2,X,Y)
    error_j_n2[i] = L2_error(u,u_e)
    plot_3D(x,y,u)
pyplot.figure(figsize=(6,6))
pyplot.grid(True)
pyplot.xlabel(r'$n_x$', fontsize=18)
pyplot.ylabel(r'$L_2$-norm of the error', fontsize=18)
pyplot.title('Error vs number of points on grid, Jacobi, n=2', fontsize = 18)
pyplot.loglog(nx_val, error_j_n2, color='k', ls='--', lw=2, marker='o')
pyplot.axis('equal');
```
